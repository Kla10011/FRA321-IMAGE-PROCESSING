```python
import numpy as np
import cv2
```

show with opencv

```python
img = cv2.imread('rick_and_morty.jpg',cv2.IMREAD_ANYCOLOR)
cv2.imshow('image',img)
k =cv2.waitKey(0)
if k == 27: # for ascii key Esc
    cv2.destroyAllWindows()
elif k == ord('s'): # for 's' key to save and exit
    cv2.imwrite('rick_only.png',img)
    cv2.destroyAllWindows()
    
```

show with matplotlib

```python
from matplotlib import pyplot as plt
img = cv2.imread('rick_and_morty.jpg',4)
plt.imshow(img, cmap='viridis', interpolation='spline36')
plt.xticks([]),plt.yticks([])
plt.show()
```
